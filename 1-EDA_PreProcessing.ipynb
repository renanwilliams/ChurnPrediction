### Pre-Processing of variables..

- Binary Categorical Variables

```python
df_final['Attrition_Flag'] = df_final['Attrition_Flag'].apply(lambda x: True if x == 'Existing Customer' else False)
```

```python
df_final['Gender'] = df_final['Gender'].apply(lambda x: 1 if x == 'F' else 0)
```

- Ordinal Categorical Variables

```python
scale_mapper = {"Uneducated": 0, "High School": 1, "College": 2,
               "Graduate": 3, "Post-Graduate": 4, "Doctorate": 5, "Unknown": 99}
df_final["Education_Level"] = df_final["Education_Level"].replace(scale_mapper)
```

```python
scale_mapper1 = {"Less than $40K": 0, "$40K - $60K": 1, "$60K - $80K": 2,
               "$80K - $120K": 3, "$120K +": 4, "Unknown": 99}
df_final["Income_Category"] = df_final["Income_Category"].replace(scale_mapper1)
```

- Nominal Categorical Variables

```python
df_final = pd.get_dummies(df_final)
```

- Correlation to identify multicollinearity

```python
corr = df_final.corr(method='spearman')
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
df_final = df_final.drop('Total_Trans_Ct',axis=1)
```
